```
import random
def interchangeSort(arr,arrC):
    n = len(arr)
    c = 1 #i=0
    for i in range(0,n-1):
        c+= 2 # <= and n-2
        c+= 2 # j =0
        for j in range(i+1, n):
            c+= 2 # j<= n-1
            c+=1 # <
            if arr[j] < arr[i]:
                c+=3 #swap 
                arr[i], arr[j] = arr[j], arr[i]
            c+=1 # j++
        c+=2 # j<=n-2
        c+=1 # i++
    c+=2 #i<=n-2
    arrC.append(c)
    print(n,":",c)

a = []


arrN = []
arrC = []

j = 794
while j <= 35000:
    arrN.append(j+1)
    for i in range(j+1):
      n = random.choice(range(10, 101))
      a.append(n)
    interchangeSort(a,arrC)
    a.clear()
    j *= 1.1
    j = int(j)
dic = {'n':arrN, 'c':arrC}
print(dic)

import pandas as pd
data = pd.DataFrame.from_dict(dic)
print(data)

data.to_csv('./data.csv', index=False)



```

```
y = np.asarray(arrC)
X = np.asarray(X)
X = X.reshape(100,1)
y = y.reshape(100,1)
plt.plot(X, y, 'ro')
plt.axis([40000, 400000000, 100000, 800000000])
plt.xlabel('n^2')
plt.ylabel('c')
plt.show()
```

```
# Building Xbar 
one = np.ones((X.shape[0], 1))
Xbar = np.concatenate((one, X), axis = 1)

# Calculating weights of the fitting line 
A = np.dot(Xbar.T, Xbar)
b = np.dot(Xbar.T, y)
w = np.dot(np.linalg.pinv(A), b)
print('w = ', w)
# Preparing the fitting line 
w_0 = w[0][0]
w_1 = w[1][0]
x0 = np.linspace(145, 185, 2)
y0 = w_0 + w_1*x0

# Drawing the fitting line 
plt.plot(X.T, y.T, 'ro')     # data 
plt.plot(x0, y0)               # the fitting line
plt.axis([100000000, 400000000, 100000000, 800000000])
plt.xlabel('n^2')
plt.ylabel('Ground count')
plt.show()
```
